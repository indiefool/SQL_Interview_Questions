```python
# Definition for singly-linked list.
class ListNode:
  def __init__(self, val=0, next=None):
    self.val = val
    self.next = next
class Solution:
    def sortList(self, head) :
        if not head or not head.next:
            return head
        temp=head
        s1=[]

        while temp:
            s1.append(temp.val)
            temp=temp.next

        s1=sorted(s1)
        head=ListNode(s1[0])
        temp=head

        for i in range (1,len(s1)):
            temp.next=ListNode(s1[i])
            temp=temp.next

        return head

l1=ListNode(4)
l1.next=ListNode(2)
l1.next.next=ListNode(1)
l1.next.next.next=ListNode(3)


s=Solution()

result=s.sortList(l1)

while result:
  print(result.val,end=' -> ')
  result=result.next

print('None')
```
